### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

**Stats of Legendary vs Non Legendary**

```python
sns.boxplot(x="Legendary", y="Total", data=df)    
sns.boxplot(x="Legendary", y="Attack", data=df)
plt.xticks([0, 1], ["Non-Legendary", "Legendary"])
plt.xlabel("Rarity")
plt.ylabel("Total Stats")
```

**Pokemon per Generation**

```python
sns.countplot(x="Generation", data=df)
plt.ylabel("No. Of Pokemon")
```

**Stats over Generations(To Identify Power Creep)**

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x="Generation", y="Total", data=df, hue="Generation", palette="Set2", legend=False)

plt.xlabel("Generation", fontsize=12)
plt.ylabel("Total Stats", fontsize=12)
sns.despine()
plt.tight_layout()
```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

```python
df["Speed"][(df["Type 1"] == "Ghost") | (df["Type 2"]=="Ghost")][df["Generation"]== 4][df["Legendary"]==False].nlargest(2).index[-1]
```

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
char_attack = df["Attack"][df.index == "Charizard"].values[0]
char_defence = df["Defense"][df.index == "Charizard"].values[0]

df[(df.Defense > char_defence) & (df.Attack < char_attack) & (df.Legendary == False)]
```

### Which type of legendary pokemons are the most common?

```python
df["Type 1"][df.Legendary ==True].value_counts().add(df["Type 2"][df.Legendary ==True].value_counts(), fill_value=0).idxmax()
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
Hp = df["HP"][(df.Generation ==6) & df.Legendary == True].min()
df[(df.Generation ==3) & (df.Legendary==False) & (df.HP > Hp) ].shape[0]
```

### Print out the third slowest pokemon(s) in the dataset.

```python
x = df.Speed.drop_duplicates().nsmallest(3).iloc[-1]
df[df.Speed==x].index
```

### Which pokemon type has the highest average HP?

```python
df.melt(id_vars="HP", value_vars=["Type 1", "Type 2"], value_name="Type").dropna(subset="Type").groupby("Type")["HP"].mean()
```
